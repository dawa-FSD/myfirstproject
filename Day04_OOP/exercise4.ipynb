```python
# exercise num 3
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.__quantity = quantity

    @property
    def quantity(self):
        return self.__quantity

    def restock(self, n):
        self.__quantity += n

    def sell(self, n):
        self.__quantity -= n


product = Product("Phone", 25000, 20)

print(product.quantity)

product.restock(10)

print(product.quantity)
```

```python
# exercise num 4
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.__quantity = quantity

    @property
    def quantity(self):
        return self.__quantity

    def restock(self, n):
        if n <= 0:
            raise ValueError("Restock amount must be positive")
        self.__quantity += n

    def sell(self, n):
        if n <= 0:
            raise ValueError("Sell amount must be positive")

        if n > self.__quantity:
            raise ValueError("Not enough stock")

        self.__quantity -= n


product = Product("Keyboard", 1200, 15)

product.sell(5)
print(product.quantity)

# product.sell(20)   # Raises ValueError
```

```python
# exercise num 5
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.__quantity = quantity

    @property
    def quantity(self):
        return self.__quantity

    def restock(self, n):
        self.__quantity += n

    def sell(self, n):
        if n > self.__quantity:
            raise ValueError("Not enough stock")
        self.__quantity -= n


p1 = Product("Laptop", 60000, 10)
p2 = Product("Mouse", 500, 50)
p3 = Product("Monitor", 15000, 8)

p1.sell(2)

print(p1.name, p1.quantity)
print(p2.name, p2.quantity)
print(p3.name, p3.quantity)
```
