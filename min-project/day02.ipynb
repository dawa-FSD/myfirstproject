```python

total = 1000
people = 4    

# Function
def split_bill(total, people, tip_rate=0.10):
    total_with_tip = total + (total * tip_rate)
    return total_with_tip / people

share = split_bill(total, people)

names = ["Dawa", "Abel", "Sara", "Lensa"]

for name in names:
    print(f"{name} pays {share:.2f} ETB")
```
