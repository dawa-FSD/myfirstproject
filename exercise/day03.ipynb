```python

#class Exercise question num 5

print("\n5. Write and Read")

names = ["Abebe", "Hana", "Dawit"]

with open("names.txt", "w") as file:
    for name in names:
        file.write(name + "\n")

with open("names.txt", "r") as file:
    for name in file:
        print(name.strip())
```

```python

#class Exercise question num 6

print("\n6. Safe Division")

try:
    number = int(input("Enter a number: "))
    result = 1000 / number
    print("Result:", result)

except ValueError:
    print("Please enter a valid number.")

except ZeroDivisionError:
    print("Cannot divide by zero.")
```

```python
#class Exercise

stock = {}

try:
    with open("stock.txt", "r") as f:
        for line in f:
            item, qty = line.strip().split(",")
            stock[item] = int(qty)

except FileNotFoundError:
    print("No stock file yet — starting empty")

def adjust(item, amount):
    stock[item] = stock.get(item, 0) + amount

adjust("Paracetamol", 5)

low = [item for item, qty in stock.items() if qty < 10]

print("Stock:", stock)
print("Low stock:", low)
```
